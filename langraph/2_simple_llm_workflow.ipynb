```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from dotenv import load_dotenv
load_dotenv()
```

```python
template = PromptTemplate(
    template='You are a helpful assistant that help to resolve user query. Please answer the following question: {question}',
    input_variables=['question']
)
```

```python
model = ChatOpenAI()
```

```python
parser = StrOutputParser()
```

```python
class conversationState(TypedDict):
    question: str
    answer: str
```

```python
def chat_qa(state: conversationState) -> conversationState:
    question = state['question']
    prompt = template.format(question=question)
    answer = model.invoke(prompt).content
    return conversationState(question=question, answer=answer)
```

```python
# Deffine the graph
graph = StateGraph(conversationState)

# Add node to your graph
graph.add_node('chat_qa', chat_qa)

# Add edges to your graph
graph.add_edge(START, 'chat_qa')
graph.add_edge('chat_qa', END)

#compile the graph
workflow = graph.compile()
```

```python
initial_state = conversationState(question='Who is bigsec', answer='')

final_state = workflow.invoke(initial_state)

print(final_state)
```
